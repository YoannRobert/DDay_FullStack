```python
import pandas as pd
```

```python
df1_meteo = pd.read_csv('./csv/open-meteo/openmeteo_paris_2000-2009.csv')
df2_meteo = pd.read_csv('./csv/open-meteo/openmeteo_paris_2010-2025.csv')
df_meteo = pd.concat([df1_meteo, df2_meteo])
df_meteo.reset_index(inplace=True, drop=True)
df_meteo['datetime'] = pd.to_datetime(df_meteo['time'])
df_meteo.head(10)
```

```python
df_meteo[ df_meteo['datetime'].dt.year == 2019 ].head()
```

```python
df_rte = pd.read_csv('./csv/power_consumption_2000_2025_hourly.csv')
df_rte['start_date'] = pd.to_datetime(df_rte['start_date'], utc=True).dt.tz_convert('Europe/Paris')
df_rte['end_date'] = pd.to_datetime(df_rte['end_date'], utc=True).dt.tz_convert('Europe/Paris')
df_rte.head()
```

```python
df_rte['temperature'] = 0
```

```python
df_rte.loc[(df_rte['start_date'].dt.year >= 2000) & (df_rte['start_date'].dt.year <= 2019), 'temperature'] = df_meteo[(df_meteo['datetime'].dt.year >= 2000) & (df_meteo['datetime'].dt.year <= 2019)]['temperature_2m']
df_rte.head()
```

```python
df_rte.to_csv('./csv/rte_paris.csv', index=False)
```
