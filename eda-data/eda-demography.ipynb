```python
import pandas as pd
```

```python
df = pd.read_csv('./csv/demographie/population_agglomeration_2023.csv')
df.head(20)
```

```python
pop_total = 66165815

pop20 = df['Population_2023'].sum()
pop10 = df['Population_2023'][:10].sum()
print(f"population des 20 plus grandes agglomérations : {pop20} habitants ({round(100*pop20/pop_total,1)}%)")
print(f"population des 10 plus grandes agglomérations : {pop10} habitants ({round(100*pop10/pop_total,1)}%)")
```
