```python
import pandas as pd
```

```python
df_power = pd.read_csv('./csv/2010_power_consumption.csv')
df_power.head(10)
```

```python
df_energy = pd.read_csv('./csv/2010_energy_consumption.csv')
df_energy['value2'] = round(df_energy['value'] / 1000)
df_energy.head(10)
```

```python
df_power['start_date'] = pd.to_datetime(df_power['start_date'], utc=True).dt.tz_convert('Europe/Paris')
df_power['end_date'] = pd.to_datetime(df_power['end_date'], utc=True).dt.tz_convert('Europe/Paris')
df_power['day_of_year'] = df_power['start_date'].dt.dayofyear
```

```python
df2 = df_power.groupby('day_of_year')['value'].sum().reset_index()
df2['value'] = round(df2['value'] / 2000)
df2.head(10)
```
