```python
import pandas as pd
```

```python
df = pd.read_csv("./csv/open-meteo/weather_historical_2000_2026_top50_reducted.csv")
print(df.shape)
```

```python
df.head()
```

```python
df['datetime_tz'] = pd.to_datetime(df['Timestamp'], utc=True).dt.tz_convert('Europe/Paris')
df.head()
```

```python
df_groupby_temp = df.groupby('datetime_tz')['Temperature'].agg([
    'mean','min','max','median','std',
    ('q1', lambda x: x.quantile(0.25)),
    ('q3', lambda x: x.quantile(0.75))
]).reset_index()
```

```python
df_groupby_temp.head()
```

```python
df_groupby_temp.to_csv("./csv/open-meteo/temperature-metrics-2020-2025.csv", index=False)
```

```python
list_cities = df['City'].unique()
print(list_cities)
print(len(list_cities))
```

```python
df2 = df[df['City'].isin(list_cities[:25])]
df2.shape
```

```python
df2_groupby_temp = df2.groupby('datetime_tz')['Temperature'].agg([
    'mean','min','max','median','std',
    ('q1', lambda x: x.quantile(0.25)),
    ('q3', lambda x: x.quantile(0.75))
]).reset_index()
```

```python
df2_groupby_temp.head()
```

```python
df_rte = pd.read_csv('./csv/power_consumption_2000_2025_hourly.csv')
df_rte['start_date'] = pd.to_datetime(df_rte['start_date'], utc=True).dt.tz_convert('Europe/Paris')
df_rte['end_date'] = pd.to_datetime(df_rte['end_date'], utc=True).dt.tz_convert('Europe/Paris')

df_rte.head()
```

```python
print(df_rte.shape)
print(df_rte.dtypes)
```

```python
print(df_groupby_temp.shape)
print(df_groupby_temp.dtypes)
```

```python
df_rte_temp = pd.merge(df_rte, df_groupby_temp, left_on='end_date', right_on='datetime_tz', how="left")
df_rte_temp.head()
```

```python
df_rte_temp.drop(columns=['datetime_tz'], inplace=True)
```

```python
df_rte_temp.head()
```

```python
df_rte_temp.to_csv("./csv/power_consumption_meteo_2000_2025_hourly.csv", index=False)
```
