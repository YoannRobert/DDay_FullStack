```python
import pandas as pd
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
os.getenv('DATA_STORAGE')
```

```python
df_rte = pd.read_csv(os.getenv('DATA_STORAGE') + 'power_consumption_meteo_2000_2025_hourly.csv')
df_rte.head()
```

```python
df_rte['start_date'] = pd.to_datetime(df_rte['start_date'], utc=True).dt.tz_convert(None)
df_rte['end_date'] = pd.to_datetime(df_rte['end_date'], utc=True).dt.tz_convert(None)
df_rte.head()
```

```python
df_cut = df_rte[df_rte['start_date'] >= '2021-01-01']
df_cut.to_csv(os.getenv('DATA_STORAGE') + 'power_consumption_meteo_2021_2025_utc.csv', index=False)
```
